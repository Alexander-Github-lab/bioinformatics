### 计算hamming distance for dna sequence in python

```python
def hamming_distance(seq1, seq2):
    # Check if the sequences have the same length
    if len(seq1) != len(seq2):
        raise ValueError("Sequences must have the same length")
    # Initialize the Hamming distance to 0
    distance = 0
    # Iterate through the sequences and compare each pair of nucleotides
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]:
            distance += 1
    return distance
```
